```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
college = pd.read_csv("College.csv", delimiter = ',')
college.head(5)
```

```python
college2 = pd.read_csv("College.csv", index_col = 0)
college2.head(5)
```

```python
college.keys()
```

```python
college3 = college.rename({'Unnamed: 0': 'College'}, axis = 1)
college3 = college3.set_index('College')
college = college3
```

```python
college.head(5)
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']], grid = True)
```

```python
college.boxplot(column = 'Outstate', by = 'Private')    
```

```python
college['Elite'] = pd.cut(college['Top10perc'], bins = 2, labels = ['No', 'Yes'])
```

```python
college.value_counts(college['Elite'])
```

```python
college.boxplot(column = 'Outstate', by = 'Elite')
```

```python
fig, ax = plt.subplots(2, 2, figsize = (6,6))
college['Apps'].plot.hist(ax=ax[0,0])
college['Accept'].plot.hist(ax=ax[0,1])
college['Enroll'].plot.hist(ax=ax[1,0])
college['Outstate'].plot.hist(ax=ax[1,1]);
```

```python
college.describe()
```

```python
college['Acp.rate'] = college['Accept'] / college['Apps']
college['Enroll.rate'] = college['Enroll'] / college['Accept']
```

```python
college.boxplot(column = 'Acp.rate', by = 'Private')
```

```python
college['Spending.student'] = college['Room.Board'] + college['Books'] + college['Personal']
```

```python
pd.crosstab(college['Private'], college['Top10perc'] > 25)
```

```python
import seaborn as sb
```

```python
sb.pairplot(college[['Top10perc', 'PhD', 'Spending.student', 'Acp.rate', 'Grad.Rate']])
```

```python
sb.jointplot(x = 'Spending.student', y = 'Grad.Rate', data = college, kind = 'reg')
```

```python
sb.kdeplot(college['Grad.Rate'])
```

```python
college.info()
```

```python
# college_norm = college.select_dtypes(include = ['int64', 'float64']).apply(lambda x: (x - x.mean()) / x.std(), axis = 1)
# college_norm = college.select_dtypes(include = ['int64', 'float64']).apply(lambda x: np.log1p(x), axis = 1)
college_norm = college.select_dtypes(include = ['int64', 'float64']).apply(lambda x: x / np.linalg.norm(x), axis = 1)
```

```python
college_norm.head(5)
```

```python
sb.pairplot(college_norm[['Spending.student', 'Grad.Rate', 'Top10perc', 'Acp.rate']])
```
